```python
# check for missing values and duplicated values
def show_details(train):
    missed_values = train.isnull().sum()
    missed_values_percent = (train.isnull().sum()) / len(train)
    duplicated_values = train.duplicated().sum()
    duplicated_values_percent = (train.duplicated().sum()) / len(train)
    info_frame = pd.DataFrame({'Missed_Values' : missed_values , 
                              'Missed_Values %' :missed_values_percent,
                              'Duplicated values' :duplicated_values,
                              'Duplicated values %':duplicated_values_percent})
    return info_frame.T
```

### Data Preprocessing: Cleaning Text

The exploratory data analysis has provided us with valuable insights into our the tweet data. Now it is time to go into the preprocessing steps and prepare the data for machine learning. The first step is cleaning the textual data within tweets themselves. We do this to remove any special characters, links, and white spaces to ensure we have clean text for tokenization and modelling steps. Again, as was mentioned in the previous section of this notebook, the code for preprocessing and developing the model is taken from: [Learn BERT - most powerful NLP algorithm by Google](https://www.udemy.com/course/bert-nlp-algorithm/learn/lecture/29758070?start=1#reviews) course.  

```python
# function to clean the tweets
def clean_tweet(tweet):
    tweet = BeautifulSoup(tweet, 'lxml').get_text() # get text from html
    tweet = re.sub(r'@[A-Za-z0-9]+', '', tweet) # remove @
    tweet = re.sub(r'https?://[A-Za-z0-9./]+', '', tweet) # remove links
    tweet = re.sub(r'[^a-zA-Z]', ' ', tweet) # remove special characters
    tweet = re.sub(r' +', ' ', tweet) # remove extra spaces
    return tweet
```

```python
# apply the function to the data, both train and validation
data_clean_train = [clean_tweet(tweet) for tweet in train['tweet']]
data_clean_val = [clean_tweet(tweet) for tweet in val['tweet']]
```

```python
# add data labels
data_labels_train = train['sentiment']
data_labels_val = val['sentiment']
```

### Data Preprocessing: Tokenization

Tokenization is the process of converting the sequence of words, such as the sentence into smaller parts known as tokens [Link](https://www.datacamp.com/blog/what-is-tokenization#). Tokens can be individual words or pairs of words. The goal of tokenization is to make the text digestible for machine learning algorithms without loosing the context. Algorithms can more easily identify patterns and learn human language in this way [Link](https://www.datacamp.com/blog/what-is-tokenization#). Because we are using BERT, we will be using its own tokanizer.        

```python
# FullTokenizer from bert library
FullTokenizer = bert.bert_tokenization.FullTokenizer
bert_layer = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/2", trainable=False)
vocab_file = bert_layer.resolved_object.vocab_file.asset_path.numpy() # get the vocab file
do_lower_case = bert_layer.resolved_object.do_lower_case.numpy()  # get the lower case boolean

tokenizer = FullTokenizer(vocab_file, do_lower_case)
```

```python
def encode_sentence(sent):
  return tokenizer.convert_tokens_to_ids(tokenizer.tokenize(sent))
```

### Creating the Model

In this section we are again following the example demonstrated in the [Learn BERT - most powerful NLP algorithm by Google](https://www.udemy.com/course/bert-nlp-algorithm/learn/lecture/29758070?start=1#reviews). 

```python
# create the layers
class DCNN(tf.keras.Model):
    def __init__(self,
                vocab_size, 
                emb_dim=128, # embedding dimension
                nb_filters=50, # number of filters
                FFN_units=512, # 512 neurons in the hidden layer
                nb_classes=4, # number of classes
                dropout_rate=0.1, 
                training=False,
                name="dcnn"):
        super(DCNN, self).__init__(name=name)
        # embedding layer
        self.embedding = layers.Embedding(vocab_size, emb_dim)
        # convolutional layers
        self.bigram = layers.Conv1D(filters=nb_filters, kernel_size=2, padding="valid", activation="relu")

        self.pool_1 = layers.GlobalMaxPool1D()

        self.trigram = layers.Conv1D(filters=nb_filters, kernel_size=3, padding="valid", activation="relu")

        self.pool_2 = layers.GlobalMaxPool1D()

        self.fourgram = layers.Conv1D(filters=nb_filters, kernel_size=4, padding="valid", activation="relu")

        self.pool_3 = layers.GlobalMaxPool1D()
        # dense layers
        self.dense_1 = layers.Dense(units=FFN_units, activation="relu")
        # dropout layer
        self.dropout = layers.Dropout(rate=dropout_rate)

        if nb_classes == 2:
            self.last_dense = layers.Dense(units=1, activation="sigmoid")
        else:
            self.last_dense = layers.Dense(units=nb_classes, activation="softmax")

    def call(self, inputs, training):
        x = self.embedding(inputs)
        x_1 = self.bigram(x)
        x_1 = self.pool_1(x_1)
        x_2 = self.trigram(x)
        x_2 = self.pool_2(x_2)
        x_3 = self.fourgram(x)
        x_3 = self.pool_3(x_3)

        merged = tf.concat([x_1, x_2, x_3], axis=-1) # (batch_size, 3 * nb_filters)
        merged = self.dense_1(merged)
        merged = self.dropout(merged, training)
        output = self.last_dense(merged)

        return output
```

After we have created the model with all the layers, we need to set hyperparameters. We will use the Adam optimizer with a learning rate of 0.0001, and the loss function will be sparse categorical crossentropy. We will train the model for 10 epochs.

```python
# hyperparameters
vocab_size = len(tokenizer.vocab)
emb_dim = 200
nb_filters = 100
FFN_units = 256
nb_classes = 4
dropout_rate = 0.1
nb_epochs = 10
```

```python
# create the model
dcnn = DCNN(vocab_size=vocab_size,
            emb_dim=emb_dim,
            nb_filters=nb_filters,
            FFN_units=FFN_units,
            nb_classes=nb_classes,
            dropout_rate=dropout_rate)
```

```python
# compile the model 
dcnn.compile(loss="sparse_categorical_crossentropy", # for multiclass classification
                 optimizer="adam",
                 metrics=["sparse_categorical_accuracy"]) # accuracy for multiclass classification
```

We will create checkpoints in case we need to recover the model if the training is interrupted. The checkpoint will restore the latest version from the checkpoint. 

```python
# checkpoint 
checkpoint_path = "./ckpt_bert_tok"
ckpt = tf.train.Checkpoint(dcnn=dcnn)
ckpt_manager = tf.train.CheckpointManager(ckpt, checkpoint_path, max_to_keep=1)

if ckpt_manager.latest_checkpoint:
    ckpt.restore(ckpt_manager.latest_checkpoint)
    print("Latest checkpoint restored!")
```

```python
class mycustomcallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        ckpt_manager.save()
        print("Checkpoint saved at {}".format(checkpoint_path))
```

```python
# train the model
dcnn.fit(dataset,
         epochs=nb_epochs,
         callbacks=[mycustomcallback()])
```

### Prediction

```python
# make predictions
def get_prediction(sentence):
    tokens = encode_sentence(sentence)
    inputs = tf.expand_dims(tokens, 0)

    output = dcnn(inputs, training=False)

    sentiment = np.argmax(output)

    if sentiment == 0:
        print("Negative")
    elif sentiment == 1:
        print("Positive")
    elif sentiment == 2:
        print("Neutral")
    else:
        print("Irrelevant")
```
